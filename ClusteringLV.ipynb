```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/content/income (2).csv')
```

```python
data.info()
```

```python
data.shape
```

```python
data.columns = ['Age', 'Income']
```

```python
x = data [['Age', 'Income']]
```

```python
data.head()
```

```python
from sklearn.preprocessing import StandardScaler #scaler - to standardize or scale (huge difference into std value)
scaler  = StandardScaler()
# to use LabelEncoder, the data shd be in panda series
# most of them or always the x-axis or features (input) shd always be in data frame
x = scaler.fit_transform(x)
```

```python
x
```

```python
# now 'x' is 2-D array
# now we want again the x to be dataFrame
x = pd.DataFrame(x)
```

```python
x
```

```python
# now we got dataframe, but without names, we have to again give names (column heading)
x.columns= ['Age', 'Income']
```

```python
x
```

```python
#Plot the points by taking x= age, y = income
plt.scatter(x['Age'], x ['Income'], color = 'red', s= 100, marker = 'o')
plt.xlabel('Age')
plt.ylabel('Income')
plt.title('Age vs Income')
plt.show()
```

```python
# The Elbow Method is used to choose the optimal number of clusters (K) in K-Means clustering.
sse =[] #sse- sum of sqaure error
from sklearn.cluster import KMeans
for k in range (2,11):
  km = KMeans(n_clusters=k, n_init= 10)
  km.fit(x)
  sse.append(km.inertia_)
```

```python
plt.scatter(range(2,11),sse,marker = 'o', color = 'red', s= 100)
plt.xlabel('Number of clusters')
plt.ylabel('sse')
plt.title('Elbow Method')
```

```python
# the transition point -> k= 3 (no.of clusters)
km = KMeans (n_clusters = 3, n_init= 10)
y = km.fit_predict(x)
```

```python
x ['clusters'] = y
```

```python
x
```

```python
y
```

```python
x.shape
```

```python
x1 = x[x['clusters']==0]
x2 = x[x['clusters']==1]
x3 = x[x['clusters']==2]
```

```python
x1
```

```python
x2
```

```python
x3
```

```python
centroids = km.cluster_centers_
```

```python
centroids
```

```python
plt.scatter(x1['Age'], x1['Income'], color= 'blue', s=300, marker = 'o')
plt.scatter(x2['Age'], x2['Income'], color= 'red', s=300, marker = '+')
plt.scatter(x3['Age'], x3['Income'], color= 'green', s=300, marker = '*')
plt.xlabel('Age')
plt.ylabel('Income')
plt.title('Cluster Plot')
plt.scatter(centriods[:,0], centriods [:,1], color='pink', s=300, marker='*')
```
